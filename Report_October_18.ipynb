###  8 -1- 4

```python
throughput_814_1 = pltme.plot_producer('8-1-4-1r/producers/producer-1/stdout-20171017-174728.csv',1,1,txt='',showPlot=False)
throughin_814 = pltme.plot_throughput_per_mini_batch('8-1-4-1r/spark-metrics-20171017-174644.csv',1,showPlot=False)
processing_time_814 =  pltme.plot_scheduling_and_Total_Delay('8-1-4-1r/spark-metrics-20171017-174644.csv',1,'title',showPlot=True)
processing_records_per_second_per_batch_814 = pltme.rec_per_sec(processing_time_814,throughin_814)
average_record_throughin = pltme.find_total_consumer_throughin('8-1-4-1r/spark-metrics-20171017-174644.csv',16*60)  #5400
print ' Through-in rate is: %d Records/sec ' % average_record_throughin

## all producers
producers = '8-1-4-1r/producers/'
adir = '/home/georgeha/repos/midas_exps/streaming/k-means/spark/1-node/8-1-4-1r/producers/'
dataset =  pltme.find_producers_file(adir)
avg_production_rate_814 = pltme.find_total_producer_throughput(producers,dataset,950,False,'')
```

```python
# create a new plot with a title and axis labels

p1 = figure(title="Figure : Avg Production Rate is:" + str(avg_production_rate_814[0])+ "Records/sec or " + str(avg_production_rate_814[0]*5000*3*8/1024**2) + " MB/s", x_axis_label='Msg Number', y_axis_label='KB/s')
p1.line(avg_production_rate_814[1], avg_production_rate_814[2], line_width=2,line_color="red")

p2 = figure(title='Spark Processing Rate in Records/sec/minibatch',x_axis_label='#miniBatch', y_axis_label='Records/sec per minibatch')
p2.line(processing_records_per_second_per_batch_814[0],processing_records_per_second_per_batch_814[1],line_width=2,line_color='purple')

p3 = figure(title='Spark throughin per minibatch - Avg ThroughIn rate is: ' + str(average_record_throughin) +' Records/sec', x_axis_label='#miniBatch', y_axis_label='Records per minibatch')
p3.line(throughin_814[0],throughin_814[1],line_color='blue',line_width=2)

p4 = figure(title = 'Figure: Spark Processing Time per minibatch', x_axis_label='#miniBatch',y_axis_label='Time in Sec')
p4.line(processing_time_814[0],processing_time_814[1],line_color='green',line_width=2)

plots = hplot(p1,p2,p3,p4)
bplt.show(plots)
```

```python
print "Processing rate is  " , processing_records_per_second_per_batch_814[1][10]
print "Number of records per minibatch ", throughin_814[1][10]
print "Time to completion of MiniBatch: ", processing_time_814[1][10]
print "Minibatch Records are: " , processing_records_per_second_per_batch_814[1][10]*processing_time_814[1][10]
#processing_records_per_second_per_batch_814[1]
```
