```python
import pandas as pd
%matplotlib notebook
pd.set_option('display.float_format', lambda x: '%.3f' % x)
import numpy as np
import matplotlib.pyplot as plt


from bokeh.plotting import figure, output_file
from bokeh.charts import Bar, output_file, show
import bokeh.plotting as bplt
from bokeh.palettes import Dark2_5 as palette
from bokeh.models import Label, Title
import itertools  
from bokeh.layouts import gridplot,row,column

from bokeh.io import hplot, output_file, show
from bokeh.plotting import figure

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
from bokeh.io import output_notebook
plt.style.use('ggplot')
output_notebook()
import plotlib as pltme
```

###  8 -1- 4

```python
throughput_814_1 = pltme.plot_producer('8-1-4-1r/producers/producer-1/stdout-20171017-174728.csv',1,1,txt='',showPlot=False)
throughin_814 = pltme.plot_throughput_per_mini_batch('8-1-4-1r/spark-metrics-20171017-174644.csv',1,showPlot=False)
processing_time_814 =  pltme.plot_scheduling_and_Total_Delay('8-1-4-1r/spark-metrics-20171017-174644.csv',1,'title',showPlot=True)
processing_records_per_second_per_batch_814 = pltme.rec_per_sec(processing_time_814,throughin_814)
```

```python
# create a new plot with a title and axis labels
p1 = figure(title="Figure : Avg Producer Throughput - Avg Production Rate: 30Records/sec", x_axis_label='Msg Number', y_axis_label='KB/s')
p1.line(throughput_814_1[0], throughput_814_1[1], line_width=2,line_color="red")

p2 = figure(title='Spark Processing Rate in Records/sec/minibatch',x_axis_label='#miniBatch', y_axis_label='Records/sec per minibatch')
p2.line(processing_records_per_second_per_batch_814[0],processing_records_per_second_per_batch_814[1],line_width=2,line_color='purple')

p3 = figure(title='Spark throughin per minibatch', x_axis_label='#miniBatch', y_axis_label='Records per minibatch')
p3.line(throughin_814[0],throughin_814[1],line_color='blue',line_width=2)

p4 = figure(title = 'Figure: Spark Processing Time per minibatch', x_axis_label='#miniBatch',y_axis_label='Time in Sec')
p4.line(processing_time_814[0],processing_time_814[1],line_color='green',line_width=2)

plots = hplot(p1,p2,p3,p4)
bplt.show(plots)
```
