## Spark Strong Scaling 

```python
## Line Plot . 

# X axis: spark nodes
x = [1,2,4,8,16]

#Y axis: total time to compute
avg_p_rate_811 = sum(processing_records_per_second_per_batch_811[1])/20

avg_p_rate_812 = sum(processing_records_per_second_per_batch_812[1])/20


avg_p_rate_814 = sum(processing_records_per_second_per_batch_814[1])/15

avg_p_rate_818 = sum(processing_records_per_second_per_batch_818[1])/17

avg_p_rate_8116 = sum(processing_records_per_second_per_batch_8116[1])/15

y = [avg_p_rate_811,avg_p_rate_812,avg_p_rate_814,avg_p_rate_818,avg_p_rate_8116]
```

```python
myDF_5k = pd.DataFrame(dict(Production_Rate=y,label=x))

p = Bar(myDF_5k,values='Production_Rate',label='label', xlabel = 'spark Nodes', ylabel="Processing Rate in Records/Sec", title='Machine: Wrangler - Spark Consumer  (strong scaling)')
show(p)
```

* In This experiment I am utilizing 1 wrangler node to produce data of total size 20GB with average data rate of 20MB/s
* I am using spark to consume and process the data via the k-means streaming application. I am increasing the number of Spark Nodes until the streaming application is stable. 

* The Appliaction is stable when I use 16 spark nodes, wheareas the throughin rate is optimal from 4 and more spark nodes. ( that is derived from the above plots) 
* Computation is still the overhead.
* I believe that the next goal should be to try to paralelize better the k-means application to increase its perfomance.
